## Get Thumbnail Pictures from URLs in the scraped Data

```python
import pandas as pd
import os
import urllib.request
import time
from urllib.error import HTTPError

path = './DataFiles/'

dataFiles = os.listdir(path)

# Working over each of the 40 sub frames in order to have 
# failsafe points from which to recover easily in case of network issues
for dataFile in dataFiles:
    
    print("Working on: {}".format(dataFile))
    dirName = 'Images/'+ dataFile.replace('.csv', '')
    
    # creating the relevant directories
    os.makedirs(dirName, exist_ok=True)
    missingLinks = []
    
    for index, row in pd.read_csv(path + dataFile, lineterminator='\n').iterrows():

        
        # some videos have been removed since creating the data frame
        # these url would send back 404 
        try:
            
            urllib.request.urlretrieve(row[2], './'+ dirName + '/' + row[0])
            
        
        except HTTPError as exception:
            
            

            print(exception)
            print('ID: '+ row[0])
            
            missingLinks.append(row[0])
            
    # All IDs where there was no thumbnail available were recorded to be later removed from the dataset
    
    with open('./'+ dirName + '/missing.txt', "w") as missingFile:

        for element in missingLinks:

            missingFile.write(element + "\n")
            
    print("Done")

```
